```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df.head()
```
